# Tratamiento de nulos:

```python
df_copia= df_crudo.copy()

df_copia["fecha"] = pd.to_datetime(df_copia["fecha"], errors="coerce")

df_copia["anio"]          = df_copia["fecha"].dt.year
df_copia["mes"]           = df_copia["fecha"].dt.month
df_copia["dia"]           = df_copia["fecha"].dt.day
df_copia["nombre_mes"]    = df_copia["fecha"].dt.month_name()
df_copia["dia_semana"]    = df_copia["fecha"].dt.day_name()
df_copia["trimestre"]     = df_copia["fecha"].dt.quarter

df_copia[["fecha", "anio", "mes", "nombre_mes", "dia", "dia_semana", "trimestre"]].head(8)

```

```python
df_copia["order_id"]=df_copia["order_id"].astype()
df_copia["cliente_id"]=df_copia["cliente_id"].astype(int)
df_copia["cantidad"]=df_copia["cantidad"].astype(int)

```
