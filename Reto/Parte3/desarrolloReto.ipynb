```python
df_copia= df_crudo.copy()

df_copia["fecha"] = pd.to_datetime(df_copia["fecha"], errors="coerce")

df_copia["anio"]          = df_copia["fecha"].dt.year
df_copia["mes"]           = df_copia["fecha"].dt.month
df_copia["dia"]           = df_copia["fecha"].dt.day
df_copia["nombre_mes"]    = df_copia["fecha"].dt.month_name()
df_copia["dia_semana"]    = df_copia["fecha"].dt.day_name()
df_copia["trimestre"]     = df_copia["fecha"].dt.quarter

df_copia[["fecha", "anio", "mes", "nombre_mes", "dia", "dia_semana", "trimestre"]].head(8)
```

# Eliminacion de datos NaN
Se tomo esta decision en base a como afectaban en las estaditicas, al ser datos que no tenia ningun espacio con valor, lo eliminamos.

```python
dfSinNan= df_copia.dropna()
dfSinNan.head()
```

```python
dfSinNan["cantidad"]= dfSinNan["cantidad"].astype(int)
dfSinNan["precio_unitario"]= dfSinNan["precio_unitario"].astype(int)
```

# Correccion dato con problema de codificacion

```python
dfSinNan['vendedor'] = dfSinNan['vendedor'].str.replace('.*Ana.*', 'Ana Lopez', regex=True)
dfSinNan.head()
```

# Crea columnas ingreso_bruto, ingreso_neto y mes. Agrupa por mes con groupby + agg

- Se define las nuevas "Variables"columnas de ingreso bruto y neto, con su respectiva operacion 

- Se hace el llamado de la culumna creada
- Se redondean las cifras y se ordena el ingreso neto de manera descendente 

```python
dfSinNan["ingresobruto"]= dfSinNan["cantidad"]*dfSinNan["precio_unitario"]
dfSinNan["ingresoneto"]= dfSinNan["ingresobruto"]*(1- dfSinNan["descuento"])

resumen_mensual=dfSinNan.groupby("mes").agg(
    Ingreso_Bruto=("ingresobruto", "sum"),
    Ingreso_Neto=("ingresoneto", "sum"),
).round().sort_values("Ingreso_Neto", ascending=False)
print(f"Resumen Mensual \n {resumen_mensual}")
```
