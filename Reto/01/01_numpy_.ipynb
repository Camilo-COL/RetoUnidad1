```python
import numpy as np
import pandas as pd

df=pd.read_csv("ventas_techventas.csv", encoding=("latin1"))
df.head()
```

```python
df.info()
```

```python
df_copy=df.copy()
cantidad=np.array(df_copy["cantidad"])
print("Matriz de cantidades")
print(cantidad)
print()
print("Matriz de precios")
precio=np.array(df_copy["precio_unitario"])
print(precio)
```

```python
precio=np.nan_to_num(precio, "nan"==1)
print(precio)
print()
cantidad=np.nan_to_num(cantidad, "nan"==1)
print()
print(cantidad)
print("Los ingresos brutos son de ",cantidad@precio)
```

```python
print("====Datos Precio====")
media=np.mean(precio)
print(f"media: {media}")
mediana=np.median(precio)
print(f"mediana: {mediana}")
desviacion=np.std(precio)
print(f"desviacion estandar: {desviacion}")
persentil=np.percentile(precio,[25,75])
print(f"percentil en 25 y 75: {persentil}")
```

```python
print("====Datos Cantidad===")
mediaCant=np.mean(cantidad)
print(f"media: {mediaCant}")
medianaCant=np.median(cantidad)
print(f"mediana: {medianaCant}")
desviCant=np.std(cantidad)
print(f"desviacion estandar: {desviCant}")
persentCant=np.percentile(cantidad,[25,75])
print(f"percentil en 25 y 75 {persentCant}")
```

# Significado de estos datos

Utilidad: Estos datos son importantes ya que nos ayudan a resumir y analizar los datos para entender una base de como se comportan en momentos especificos, la media y mdiana son los datos mas centrales/promedio, siempre y cuando sean datos sin outliers, la desviacion estandar nos muestra que tanto estan por encima o por debajo los datos con respecto a nuestro estandar y los percentiles son el porcentaje de datos que estan por debajo de nuestro estandar, todos estos datos ayudan a la empresa a tener un panorama de como se comportan sus datos y puede indicar ciertas tendecias, sesgos, oportunidades, alzas o bajas.
